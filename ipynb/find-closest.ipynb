```python
# Group the output file and find the closest location from every point
# Teemu Kakko, September 2019

import pandas as pd
df = pd.read_csv('OUTPUT_FILE.csv')

df.head() # See the first 5 rows

df.groupby(['Id', 'name']).first() # Group by Id and name

# Leave only that 'name' that has the smallest 'tt' value
grouped = df.loc[df.groupby('Id')['tt'].idxmin()]
grouped.reset_index().to_csv('FINAL_OUTPUT_FILE.csv')
```
