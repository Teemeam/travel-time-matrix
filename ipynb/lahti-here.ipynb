```python
# Create a travel time matrix using the Digitransit Routing API
# Teemu Kakko, September 2019

import requests, json
import pandas as pd
import numpy
import csv

# Origin and target coordinates
origins = pd.read_csv('ORIGIN_COORDINATES.csv')
targets = pd.read_csv('TARGET_COORDINATES.csv')

# Create an array for the results
array = numpy.array([['name','Id','origin_latitude','origin_longitude','target_latitude','target_longitude','tt']])

# Progress bar
rounds = 1

# For-loop for origin coordinates
counter = 0
for counter in range(counter,len(origins.index)):
    
    # For-loop for target coordinates
    another_counter = 0
    for another_counter in range(another_counter,len(targets.index)):
  
        # API URL
        url = 'https://matrix.route.api.here.com/routing/7.2/calculatematrix.json?'

        # HERE app id
        app_id = '823cvYz7v7ACeImIPWFC'

        # HERE app code
        app_code = 'rQTIrKdQ5mzT9oTwLmigrw'

        # Origin coordinate
        start0 = 'geo!' + origins.iloc[counter]['latitude'].replace(',','.') + ',' + origins.iloc[counter]['longitude'].replace(',','.')

        # Target coordinate
        destination0 = 'geo!' + targets.iloc[another_counter]['latitude'].replace(',','.') + ',' + targets.iloc[another_counter]['longitude'].replace(',','.')

        # Mode
        mode = 'fastest;car;traffic:enabled'

        # Departure time
        departure = 'now'

        # Add travel time to results
        summaryAttributes = 'traveltime'
  
        # GET
        r = requests.get(url + 'app_id=' + app_id + '&app_code=' + app_code + '&start0=' + start0 + '&destination0=' + destination0 + '&mode=' + mode + '&departure=' + departure + '&summaryAttributes=' + summaryAttributes) 
  
        # Extracting data in json format 
        data = r.json() 

        # Set results to variables
        name = origins.iloc[counter]['name'] # Name of the shopping centre
        Id = targets.iloc[another_counter]['Id'] # Id of the target rectangle
        origin_latitude = origins.iloc[counter]['latitude'].replace(',','.') # Origin latitude
        origin_longitude = origins.iloc[counter]['longitude'].replace(',','.') # Origin longitude
        target_latitude = targets.iloc[another_counter]['latitude'].replace(',','.') # Target latitude
        target_longitude = targets.iloc[another_counter]['longitude'].replace(',','.') # Target longitude
        tt = data['response']['matrixEntry'][0]['summary']['travelTime'] # Travel time
        
        # Add row to array
        array = numpy.append(array, [[name, Id, origin_latitude, origin_longitude, target_latitude, target_longitude, tt]], axis = 0)
    
        # Print progress
        print('Rounds: ' + str(rounds) + ' out of ' + str(len(origins.index) * len(targets.index)))
        rounds += 1

# Write rows to csv
with open('OUTPUT_FILE.csv', 'w') as csv_output:
    writer = csv.writer(csv_output)
    for row in array:
        writer.writerow(row)
```
